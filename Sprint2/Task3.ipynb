Вариант задания № 7
Описание задачи:
Создать базовый класс BankAccount в C#, который будет представлять
информацию об учетных записях в банке. На основе этого класса разработать 2-3
производных класса, демонстрирующих принципы наследования и полиморфизма.
В каждом из классов должны быть реализованы новые атрибуты и методы, а также
переопределены некоторые методы базового класса для демонстрации
полиморфизма.
Требования к базовому классу BankAccount:
• Атрибуты: Номер счета (AccountNumber), Баланс (Balance), Тип счета
(AccountType).
• Методы:
o GetInfo(): метод для получения информации о счете в виде строки.
o Deposit(decimal amount): метод для внесения денег на счет.
o Withdraw(decimal amount): метод для снятия денег со счета.
Требования к производным классам:
1. Сберегательный счет (SavingsAccount): Должен содержать дополнительные
атрибуты, такие как Процентная ставка (InterestRate). Метод Deposit() должен
быть переопределен для добавления процентов к сумме вклада при
внесении денег на счет.
2. Текущий счет (CheckingAccount): Должен содержать дополнительные
атрибуты, такие как Лимит овердрафта (OverdraftLimit).
Метод Withdraw() должен быть переопределен для проверки и применения
лимита овердрафта при снятии денег со счета.
3. Инвестиционный счет (InvestmentAccount) (если требуется третий класс):
Должен содержать дополнительные атрибуты, такие как Список активов
(AssetsList). Метод GetInfo() должен быть переопределен для включения
информации о списках активов в описании счета.

```
public interface IBankAccount
{
    void Puch();
}

public interface IExchangerate
{
    void INeedDollar();  
}

public interface IDelays
{
    void Delays(int day);
}

public class BankAccount
{
    
    public string _accountNumber;
    public int _balance;
    public string _accountType; //тип счета

    //Конструктор
    public BankAccount(string accountNumber, int balance, string accountType)
    {
        _accountNumber = accountNumber;
        _balance = balance;
        _accountType = accountType;
    }

    // Геттеры сеттеры
    public string AccountNumber
    {
        get{return _accountNumber;}
        set{_accountNumber = value;}
    }
   
    public int Balance
    {
        get{return _balance;}
        set{
            if (value >= 0)
                _balance = value;
            else
                throw new ArgumentOutOfRangeException("Баланс не может быть отрицательной!");
            }
    }

    public string AccountType
    {
        get{return _accountType;}
        set{_accountType = value;}
    }
    
    public virtual void Delays(string day)
    {
        Console.WriteLine($"{day} просрочка");
    }
    //Обычный метод
    public virtual string GetInfo()
    {
        return $"Номер аккаунта: {AccountNumber}, Баланс (в рублях): {Balance}, Тип счета: {AccountType}";
    }

    public virtual void Deposit(decimal amount)
    {
        Console.WriteLine($"Выполнена операция: Cчет № {AccountNumber} поступление средств: {amount}.");
    }

    public virtual void Withdraw(decimal amount)
    {
        Console.WriteLine($"Выполнена операция: Cчет № {AccountNumber} вывод средств:  {amount}.");
    }

    //Метод взаимодействя
    public void tranAkNumber(BankAccount akNumber, int summa)
    {
        Balance -= summa;
        akNumber.Balance += summa;
        Console.WriteLine($"Пришел перевод от {AccountNumber} на {akNumber.AccountNumber} сумма {summa}");
    }

    
}

public class SavingsAccount: BankAccount, IBankAccount// Множественное наследование 
{
    private double _interestRate;//Процентная ставка

    public SavingsAccount(string accountNumber, int balance, string accountType, double interestRate) : base (accountNumber, balance, accountType)
    {
        AccountNumber = accountNumber;
        Balance = balance;
        AccountType = accountType;
        InterestRate = interestRate;
    }

    public double InterestRate
    {
        get { return _interestRate; }
        set { _interestRate = value; }
    }

    public override void Deposit(decimal amount)
    {
        Console.WriteLine($"Процентная ставка Сбербанка №{AccountNumber}, на сумму {amount} руб. составляет {InterestRate} %."); 
    }

    public void Puch() 
    {
        Console.WriteLine("Пуш Уведомление!");
    }
}

public class CheckingAccount: BankAccount , IDelays// Простое наследование
{
    public int _overdraftLimit;

    public CheckingAccount(string accountNumber, int balance, string accountType, int overdraftLimit, int day, int percent) : base (accountNumber, balance, accountType)
    {
        OverdraftLimit = overdraftLimit;
        AccountNumber = accountNumber;
        Balance = balance;
        AccountType = accountType;

    }
    public override void Delays(string day)
    {
       Console.WriteLine($"Просрочка по кредиту {day}");

    }
    public void Delays(int day)
    {
        Console.WriteLine(day);
    }
    
    public int OverdraftLimit
    {
        get { return _overdraftLimit; }
        set { 
            
             if (value >= 0)
                _overdraftLimit = value;
            else
                throw new ArgumentOutOfRangeException("Овердрафт не может быть отрицательным!");
            
            _overdraftLimit = value; }
    }

    public override void Withdraw(decimal amount)
    {
        Console.WriteLine($"{AccountNumber}, имеет лимит овердрафта: {OverdraftLimit}.");
    }
}

public class InvestmentAccount: SavingsAccount, IExchangerate//Сложное наследование
{
    public string _assetsList;
    public int Dollar;

    public InvestmentAccount(string accountNumber, int balance, string accountType,  double interestRate, string assetsList): base (accountNumber, balance, accountType, interestRate)
    {
        AssetsList = assetsList;
        AccountNumber = accountNumber; 
        Balance = balance;
        AccountType = accountType;
        InterestRate = interestRate;
       
    }

    public string AssetsList
    {
        get { return _assetsList; }
        set { _assetsList = value; }
    }

    public override string GetInfo()
    {
        return $"Номер аккаунта: {AccountNumber}, Баланс (в рублях): {Balance}, Тип счета: {AccountType},\nДоля всех активов {InterestRate}% Список активов: {AssetsList}"; 
    }
    
    public void INeedDollar()
    {
        Dollar = base.Balance / 99;
        Console.WriteLine($"Баланс в Баксах: {Dollar}");
    }
}

BankAccount GoogeRA = new BankAccount("+7(904)472-84-52", 10544, "Физическое лицо.");
BankAccount Vlad = new BankAccount("+7(922)472-84-45", 2544, "Юр. лицо."); //объект взаимодействия 
Console.WriteLine(GoogeRA.GetInfo());
GoogeRA.Deposit(1500);
GoogeRA.Withdraw(5000);
GoogeRA.tranAkNumber(Vlad, 5000); //метод взаим
Console.WriteLine(GoogeRA.GetInfo());

Console.WriteLine("");

SavingsAccount SaveBalanse = new SavingsAccount("234567436745745", 25000, "Накопительный счет.", 8.9);
Console.WriteLine(SaveBalanse.GetInfo());
SaveBalanse.Deposit(25000);
SaveBalanse.Puch();

Console.WriteLine("");

CheckingAccount Credit = new CheckingAccount("2222-4455-3242-5252", 25000, "Крдитная карта.", 530000, 12, 12);
Console.WriteLine(Credit.GetInfo());
Credit.Withdraw(15000);
Credit.Delays("двенадцать дней");
Credit.Delays(12);

 
Console.WriteLine("");

InvestmentAccount Assets = new InvestmentAccount("563475382532523", 5000, "Инвестиционный счет.", 8.8 , "Недвижемость.");
Console.WriteLine(Assets.GetInfo());
Assets.INeedDollar();


```
